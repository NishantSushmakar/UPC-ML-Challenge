```python
encoder = OneHotEncoder()
encoded_array = encoder.fit_transform(df[['language']])
encoded_df = pd.DataFrame(encoded_array.toarray(), columns=encoder.get_feature_names_out(['language']))
```

```python
family_encoder = OneHotEncoder()
encoded_array_familiy = family_encoder.fit_transform(df[['language_group']])
encoded_family_df = pd.DataFrame(encoded_array_familiy.toarray(),columns=family_encoder.get_feature_names_out(['language_group']))
```

```python
final_df = pd.concat([df,encoded_df,encoded_family_df],axis=1)
```
